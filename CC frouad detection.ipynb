```python
x=df.drop(columns=['Class'],axis=1)
y=df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_scaler = sc.fit_transform(x)
```

```python
x_scaler[-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaler,y ,test_size=0.25,random_state=40,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,f1_score,classification_report
```

```python
# Testing
y_pred = lr.predict(x_test)
con_mat = confusion_matrix(y_test,y_pred)
print("Confusion Matrix :\n",con_mat)
accuracy_test = f1_score(y_test,y_pred)
print(" Accuracy :",accuracy_test)
clf_report = classification_report(y_test,y_pred)
print("Classication Report : \n",clf_report)
```

```python
# Trainig
y_pred = lr.predict(x_train)
con_mat = confusion_matrix(y_train,y_pred)
print("Confusion Matrix :\n",con_mat)
accuracy_train = f1_score(y_train,y_pred)
print(" Accuracy :",accuracy_train)
clf_report = classification_report(y_train,y_pred)
print("Classication Report : \n",clf_report)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
# Testing
y_pred = rf.predict(x_test)
con_mat = confusion_matrix(y_test,y_pred)
print("Confusion Matrix :\n",con_mat)
accuracy_test = f1_score(y_test,y_pred)
print(" Accuracy :",accuracy_test)
clf_report = classification_report(y_test,y_pred)
print("Classication Report : \n",clf_report)
```

```python
# Trainig
y_pred = rf.predict(x_train)
con_mat = confusion_matrix(y_train,y_pred)
print("Confusion Matrix :\n",con_mat)
accuracy_train = f1_score(y_train,y_pred)
print(" Accuracy :",accuracy_train)
clf_report = classification_report(y_train,y_pred)
print("Classication Report : \n",clf_report)
```
